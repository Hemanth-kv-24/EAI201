```python
# Create a new column called winner
filtered_df['winner'] = filtered_df.apply(lambda row: 
                                          row['home_team'] if row['home_score'] > row['away_score'] 
                                          else (row['away_team'] if row['away_score'] > row['home_score'] 
                                                else 'Draw'), axis=1)
```

```python
# Create a new numeric result column
filtered_df['result_numeric'] = 0   # default draw

filtered_df.loc[filtered_df['home_score'] > filtered_df['away_score'], 'result_numeric'] = 1    # home win
filtered_df.loc[filtered_df['home_score'] < filtered_df['away_score'], 'result_numeric'] = -1   # away win
```
